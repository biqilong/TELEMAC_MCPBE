# Example of 3d plots

Examples of 3d plots

```python
from os import environ, path
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from data_manip.extraction.telemac_file import TelemacFile
from postel.plot3d import plot3d_scalar_map
```

## Importing Telemac results

```python
# File we are going to use
file_name = path.join(environ['HOMETEL'], 'examples', 'telemac2d', 'gouttedo', 'f2d_gouttedo.slf')

# Opening the file with TelemacFile
res = TelemacFile(file_name)

# Getting water depth value at time 2
data = res.get_data_value('WATER DEPTH', 2)
```

## 3d plot of a scalar using data as z coordinates

```python
# Initialisaing matplotlb figure
fig = plt.figure(figsize=(12,7))

# Adding a 3d axe
axe = Axes3D(fig)

#Plotting mesh
plot3d_scalar_map(fig, axe, res.tri, data, x_label='X (m)', y_label='Y (m)', data_name='water depth (m)')

# Sohowing figure
plt.show()

del res
```
