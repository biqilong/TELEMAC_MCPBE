# Difference between two SERAFIN files

Two tools are available do check the difference between two SERAFIN files:
  * A comparaison of what is different between the two files
  * Creating  a new file that contains for each variable each time step data from file 1 - data from file2
  
## Diff information

For that we use the script **diff_selafin.py**.

```python
%%bash

diff_selafin.py --help
```

```python
%%bash

cd $HOMETEL/examples/telemac2d/gouttedo

diff_selafin.py f2d_gouttedo.slf f2d_gouttedo_cin.slf --head --core
```

## Diff in a file

For that we use the script **run_selafin.py diff**.

The files must have the same mesh.

For the variable and the time steps it will only keep the ones that are only in both file.

```python
%%bash

run_selafin.py diff -h
```

```python
%%bash

cd $HOMETEL/examples/telemac2d/malpasset

run_selafin.py diff f2d_malpasset-char.slf f2d_malpasset-cin1.slf diff.slf

run_selafin.py scan --core f2d_malpasset-char.slf > file1.log
run_selafin.py scan --core f2d_malpasset-cin1.slf > file2.log

run_selafin.py scan --core diff.slf

colordiff file1.log file2.log
```
