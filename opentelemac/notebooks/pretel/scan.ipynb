# Scanning a Telemac file

## Scanning a result/mesh file

The following command will write information on your files.
Such as :
- The number of elements, nodes, boundary points , planes
- List of variables 
- Number of timesteps
- Min/Max for each variable at each timestep

```python
%%bash
run_selafin.py scan -h
```

### Simple scan with just mesh information

The command below will only display mesh information

```python
%%bash
run_selafin.py scan $HOMETEL/examples/telemac2d/gouttedo/f2d_gouttedo.slf
```

### Simple scan with just variable information

The command below will be the same as below but with also the information for each variable for each time step

```python
%%bash
run_selafin.py scan --core $HOMETEL/examples/telemac2d/gouttedo/f2d_gouttedo.slf
```

## Scanning a Tomawac spectrum file

A specific scan is avaialble for spectrum files

```python
%%bash
run_selafin.py spec -h
```

### Simple scan

The basic scan.

```python
%%bash
run_selafin.py spec $HOMETEL/examples/tomawac/turning_wind/spe_ref.slf
```

### Scan with varaible information

The full scan.
You can use the option --accuracy=int to specify the number of figures to display for the variable values.

```python
%%bash
run_selafin.py spec --core $HOMETEL/examples/tomawac/turning_wind/spe_ref.slf
```
